## Classification

Multiple algorithms would be tried with our data.

- Decision Trees
- Random Forest
- Support Vector Machines
- Support Vector Machines (Kernel based)

## Model Evaluation

We evaluate our models based on __Accuracy, Precision, Recall and F1-Score__.

__The predict_proba method gives the probability of each team's winning__

### K-Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
classifier.fit(train, y_train)

# Predicting the Test set results
y_pred = classifier.predict(test)

print("Probability predictions for which team would win each match up in order for the test data")
print(classifier.predict_proba(test))

p, r, f1, _ = precision_recall_fscore_support(y_test, y_pred, average='micro')
acc = accuracy_score(y_test, y_pred)
print("Accuracy: ", acc)
print("Precision: ", p)
print("Recall: ", r)
print("F1 Score: ", f1)

# Making the Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

### Decision Tree Classifier

```python
# Fitting Decision Tree Classification to the Training set
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
classifier.fit(train, y_train)

y_pred = classifier.predict(test)

p, r, f1, _ = precision_recall_fscore_support(y_test, y_pred, average='micro')
acc = accuracy_score(y_test, y_pred)
print("Accuracy: ", acc)
print("Precision: ", p)
print("Recall: ", r)
print("F1 Score: ", f1)

cm = confusion_matrix(y_test, y_pred)
cm
```

### Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 25, criterion = 'entropy', random_state = 0)
classifier.fit(train, y_train)

# Predicting the Test set results
y_pred = classifier.predict(test)

p, r, f1, _ = precision_recall_fscore_support(y_test, y_pred, average='micro')
acc = accuracy_score(y_test, y_pred)
print("Accuracy: ", acc)
print("Precision: ", p)
print("Recall: ", r)
print("F1 Score: ", f1)
# print(classifier.predict_proba(train))

# Making the Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

### Support Vector Machines

```python
from sklearn.svm import SVC
classifier = SVC(kernel = 'linear', random_state = 0, probability=True)
classifier.fit(train, y_train)

# Predicting the Test set results
y_pred = classifier.predict(test)


p, r, f1, _ = precision_recall_fscore_support(y_test, y_pred, average='micro')
acc = accuracy_score(y_test, y_pred)
print("Accuracy: ", acc)
print("Precision: ", p)
print("Recall: ", r)
print("F1 Score: ", f1)

# Making the Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

### Support Vector Machines (RBF kernel)

```python
from sklearn.svm import SVC
classifier = SVC(kernel = 'rbf', random_state = 0, probability=True)
classifier.fit(train, y_train)

# Predicting the Test set results
y_pred = classifier.predict(test)

# print(classifier.predict_proba(test))

p, r, f1, _ = precision_recall_fscore_support(y_test, y_pred, average='micro')
acc = accuracy_score(y_test, y_pred)
print("Accuracy: ", acc)
print("Precision: ", p)
print("Recall: ", r)
print("F1 Score: ", f1)

# Making the Confusion Matrix

cm = confusion_matrix(y_test, y_pred)
cm
```
